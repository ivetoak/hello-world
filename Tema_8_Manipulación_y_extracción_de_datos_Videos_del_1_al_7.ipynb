```python
# genero valores random y numeros enteros (randint) del 0 al 10 y que las dimensiones de mi array
# sean 3 filas, 3 columnas, 3 profundidad,

array_a = np.random.randint(0, 10, (3, 3, 3))
print(array_a)
```



#### Tipos de suma de variables de una matriz (Nativos Python y Nativos Numpy)



```python
# La suma se hace el 1,1 de la primera matriz con el 1,1 de la segunda con el 1,1 de la tercera

sum(array_a)
```

```python
# Si quiero una suma total de todos los elementos hago uan suma de suma. El tensor matriz se suma a si mismo y suna todos los valores de la columna 1,
#luego la 2 y por último la 3

sum(sum(array_a))
```

```python
# Si quiero el valor total de todos, vuelvo a sumar a la suma de la suma. Estas serian funciones nativas de python
sum(sum(sum(array_a)))
```

```python
# Pero la libreria numpy ya viene con una función de suma de todas las variables del tensor. Esta es una funcion nativa de Numpy.
np.sum(array_a)
```

```python
# Numpy también tiene suma acumulativa. És una función de agregación que primero muestra el primer valor de la primera matriz (posición 1,1,1), luego
# suma la primera variable (posición 1,1,1) con la segunda (posición 2,1,1),
# luego suma esa suma de (posición 1,1,1) + (posición 2,1,1) con la posición (posición 3,1,1) y así sucessivamente.
np.cumsum(array_a)
```



#### Suma entre matrices

**Tipos de datos**

Pandas dispone de dos estructuras de datos diferentes


*   Series: estructuras de una dimensión

Estas estructuras se construyen a partir de arrays de la librería NumPy, añadiendo nuevas funcionalidades





**Serie**

Son estructuras similares a los arrays de una dimensión. Són homogéneas, es decir, sus elementos tienen que ser del mismo tipo, y su tamaño es inmutable, es decir, no se puede cambiar aunque si su contenido.

Un DataFrame contiene dos indices, uno para las filas y otro para las columnas, y se puede acceder a sus elementos mediante los nombres de las filas y las columnas.

## Video 7. Merge & Concatenación




### Merge

Merge es lo mismo que los Joins en SQL. Uniones de tablas que podemos hacer a raíz de condiciones.

Joins SQL.png

```python
# Generamos 2 DataFrames. Uno con ID y Nombre y otro ID y Edad. 2 pares clave:valor=diccionario cada uno generando
# en cada una de ellas 2 columnas y 5 filas

# Primero generamos las variables: data1 y data 2
# Luego a esas variables las mencionamos DataFrames de pandas (pd): df1 y df2

data1 = {'ID': [1, 2, 3, 4],
         'Nombre': ['Alice', 'Bob', 'Charlie', 'David']
}

df1 = pd.DataFrame(data1)

data2 = {'ID': [2, 3, 4, 5],
         'Edad': [25, 30, 35, 40]
}

df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# Inner. Solo unimos los datos presentes en los 2 Dataframes
# Podemos hacerlo indicando la clave de unión, en este caso 'ID'

inner_df = pd.merge(df1, df2, on = 'ID')
inner_df
```

```python
# Inner. Solo unimos los datos presentes en los 2 Dataframes
# O, indicando la función inner donde encontrará la clave de unión por si solo.

inner_df = pd.merge(df1, df2, how = 'inner')
inner_df
```

```python
# Left. En el que aquí sí que debemos aclarar el DataFrame que destacarà frente al otro.
# Resultando todos los valores del df1 + las columnas coincidentes del df2 segun el ID.
# El valor que no tiene lo rellena con un nulo --> (Edad ID 1 = NaN)

left_df = pd.merge(df1, df2, on = 'ID', how = 'left')
left_df
```

```python
# Right. En el que aquí sí que debemos aclarar el DataFrame que destacarà frente al otro.
# Resultando todos los valores del df2 + las columnas coincidentes del df1 segun el ID.
# El valor que no tiene lo rellena con un nulo --> (Nombre ID 5 = NaN)

right_df = pd.merge(df1, df2, on = 'ID', how = 'right')
right_df
```

```python
# Full Join = Outer. En el que unimos las 2 tablas.
# Resultando todos los ID, del 1 al 5
# Los valores que no tiene los rellena con un nulo --> (Edad ID 1 = NaN) & (Nombre ID 5 = NaN)

outer_df = pd.merge(df1, df2, on = 'ID', how = 'outer')
outer_df
```

### Concatenación de DataFrames

```python
# Concatenación de DataFrames. Para unir una tabla con otra de manera horizontal o vertical.

data1 = {'A': [1, 2, 3], 'B': [4, 5, 6]}
data2 = {'A': [7, 8, 9], 'B': [10, 11, 12]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# De manera vertical. Es decir, añadiendo filas, por lo tanto, se concatenan por columnas.

result_vertical = pd.concat([df1, df2], axis=0)
result_vertical
```

```python
# De manera horizontal. Es decir, añadiendo columnas, por lo tanto se concatenan por filas.

result_vertical = pd.concat([df1, df2], axis=1)
result_vertical
```
