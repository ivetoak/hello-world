### <font color='red'> 1. Ejercicio estructuras de control con contador <font color='black'>


Imagina que trabajas en el departamento de servicio al cliente de una compa√±√≠a que acaba de lanzar un nuevo producto y has recibido una serie de comentarios de los clientes. Quieres analizar r√°pidamente estos comentarios para ver cu√°ntos mencionan un t√©rmino espec√≠fico relacionado con el producto, como podr√≠a ser el nombre del producto.

- Queremos analizar una lista de comentarios de clientes sobre un nuevo producto y contar cu√°ntos de estos comentarios mencionan un t√©rmino espec√≠fico relacionado con el producto:

  1. Define una variable llamada **nombre_producto** que contenga el nombre del producto que deseas buscar en los comentarios:

  > EcoFriendlyBottle

  2. Define una *lista* llamada **comentarios** que contenga varios comentarios de clientes sobre el producto.

  > "Realmente amo mi nueva EcoFriendlyBottle, ¬°es perfecta para llevar al trabajo!"

  > "No estoy seguro de si la EcoFriendlyBottle vale su precio."

  > "¬øAlguien ha probado la EcoFriendlyBottle? Estoy pensando en comprar una."

  > "Preferir√≠a algo m√°s colorido que una EcoFriendlyBottle."

  > "¬°Esta botella mantiene mi bebida fr√≠a por horas!"

  3. Utiliza un bucle **for** para iterar sobre cada comentario en la lista comentarios.

  4. Dentro del bucle, verifica si el nombre_producto aparece dentro del comentario.

  5. Si el nombre_producto se encuentra en el comentario, incrementa un contador que lleva el registro de las menciones encontradas.
  
  6. Imprime cada comentario que mencione el nombre_producto.
  
  7. Al final, imprime el total de menciones encontradas del nombre_producto.

````python
print("-"*65)
print("--- LA VARIABLE SE ENCUENTRA 1 SOLA VEZ POR ELEMENTO ---")
print("-"*65)

# Se define una variable de cadena con el nombre del producto.
nombre_producto = "EcoFriendlyBottle"

# Se define la lista 'comentarios' con varios comentarios de clientes.
comentarios = [
    "Realmente amo mi nueva EcoFriendlyBottle, ¬°es perfecta para llevar al trabajo!",
    "No estoy seguro de si la EcoFriendlyBottle vale su precio.",
    "¬øAlguien ha probado la EcoFriendlyBottle? Estoy pensando en comprar una.",
    "Preferir√≠a algo m√°s colorido que una EcoFriendlyBottle.",
    "¬°Esta botella mantiene mi bebida fr√≠a por horas!"
]

# Contador para los comentarios que mencionan el nombre del producto.
print("--- COMENTARIOS DONDE SE ENCUENTRA 'EcoFriendlyBottle' ---")
print("-"*65)
contador = 0

# Utiliza el bucle 'for' para iterar sobre cada comentario.
for comentario in comentarios:

# Verifica si el nombre_producto aparece dentro del comentario y printealo.
  if nombre_producto in comentario:
    contador += 1
    print(f"{contador}¬∫ comentario que menciona el producto: {comentario}")

# Imprime el total de menciones encontradas.
print("-"*65)
print("--- RESULTADO FINAL 1 ---")
print("-"*65)
print(f"Total de menciones encontradas de '{nombre_producto}': {contador}")
print("-"*65)
````

````python
print("-"*65)
print("--- LA VARIABLE PUEDE ENCONTRARSE + DE 1 VEZ POR ELEMENTO ---")
print("-"*65)

# Se define una variable de cadena con el nombre del producto.
nombre_producto_2 = "EcoFriendlyBottle"

# Se define la lista 'comentarios_2' con varios comentarios de clientes.
comentarios_2 = [
    "Realmente amo mi nueva EcoFriendlyBottle, ¬°EcoFriendlyBottle, es perfecta para llevar al trabajo!",
    "No estoy seguro de si la EcoFriendlyBottle vale su precio.",
    "¬øAlguien ha probado la EcoFriendlyBottle? Estoy pensando en comprar una.",
    "Preferir√≠a algo m√°s colorido que una EcoFriendlyBottle.",
    "¬°Esta botella mantiene mi bebida fr√≠a por horas!"
]

# Contador para el total de menciones del nombre del producto.
contador_total_menciones = 0

print("--- COMENTARIOS DONDE SE ENCUENTRA 'EcoFriendlyBottle' Y N√öMERO DE MENCIONES ---")
print(("-"*65), "\n")

# Utiliza el bucle 'for' para iterar sobre cada comentario.
for i, comentario_2 in enumerate(comentarios_2):

# 2. Usa el m√©todo count() en el comentario para saber cu√°ntas veces aparece el producto en 1 comentario.
  menciones_en_comentario = comentario_2.count(nombre_producto_2)

# 3. Suma el resultado al contador_total_menciones.
  contador_total_menciones += menciones_en_comentario

# Verifica si el nombre_producto aparece dentro del comentario (menciones_en_comentario > 0).
  if menciones_en_comentario > 0:

# Adjusted print statement to use 'i' and 'comentario_2'
# Verifica si el nombre_producto aparece dentro del comentario y printealo.
    print(f"Comentario #{i+1}: '{comentario_2}' -> {menciones_en_comentario} menci√≥n(es) encontrada(s).")
print("\n", ("-"*65))
print("--- RESULTADO FINAL 2 ---")
print("-"*65)
# 5. Imprime el total de menciones encontradas.
print(f"\nTotal de menciones encontradas de '{nombre_producto_2}': {contador_total_menciones}\n")
print("-"*65)
````

### <font color='red'> 2. Ejercicio usuarios

Supongamos que est√°s trabajando con una lista de nombres de usuarios que se han recopilado a trav√©s de un formulario en l√≠nea. A menudo, los usuarios pueden ingresar accidentalmente espacios adicionales antes o despu√©s de sus nombres, lo que podr√≠a causar problemas al realizar comparaciones o al almacenar estos datos en una base de datos.

- Se te proporciona una lista llamada **lista_usuarios** que incluye nombres de usuarios con y sin espacios extra al principio o al final.

> "Ana", "Carlos Andres  ", " Diana", " Jorge    "

- Tu tarea es limpiar los espacios al principio y al final de cada nombre de usuario en la lista utilizando la comprensi√≥n de listas y el m√©todo .strip().

- Despu√©s de la limpieza, debes imprimir cada nombre de usuario para mostrar la eliminaci√≥n de los espacios.

- Adem√°s, proporciona un ejemplo adicional que compare la longitud de un nombre de usuario antes y despu√©s de la limpieza para demostrar la eliminaci√≥n de espacios.






````python
# Se define una lista que incluye nombres de usuarios con y sin espacios extra al principio o al final.

lista_usuarios = ["Ana", "Carlos Andres ¬†", " Diana", " Jorge ¬† ¬†"]

# Limpieza de espacios usando .strip
lista_usuarios_limpia = [usuario.strip() for usuario in lista_usuarios]

# Comparaci√≥n Original vs. Limpio

print("-"*65)
print("--- COMPARACI√ìN POR USUARIO (ORIGINAL vs. LIMPIO) ---")
print("-"*65)
print("√çndice | Original (Long.) | Limpio (Long.) | Diferencia")
print(("-"*65), "\n")

# Usamos zip() para iterar sobre ambas listas al mismo tiempo
for i, (usuario_original, usuario_limpio) in enumerate(zip(lista_usuarios, lista_usuarios_limpia)):

    len_original = len(usuario_original)
    len_limpio = len(usuario_limpio)
    diferencia = len_original - len_limpio

# Dentro del bucle for, imprimimos la informaci√≥n del usuario original y limpio en la misma l√≠nea
    print(f"[{i}]")
    print(f"  Original: '{usuario_original}' (Longitud: {len_original})")
    print(f"  Limpio: ¬† '{usuario_limpio}' (Longitud: {len_limpio})")
    print(f"  Ahorro: ¬† {diferencia} caracteres\n")

print("-"*65)
````

### <font color='red'> 3. Ejercicio votaci√≥n pel√≠culas

Imag√≠nate organizando una noche de cine con amigos y familiares, donde todos tienen la oportunidad de votar por su pel√≠cula favorita de una lista preseleccionada.

El objetivo de este desaf√≠o es construir un sistema simple pero completo en Python que no solo permita registrar esos votos, sino tambi√©n analizarlos e informar sobre la pel√≠cula m√°s popular

- **Listado de Pel√≠culas:** Comienza mostrando las pel√≠culas disponibles para votaci√≥n, enumeradas para facilitar la elecci√≥n del usuario (usar **enumerate**)

```enumerate(iterable, start=0)```

- **Registrar Votos:** Se invita al usuario a votar introduciendo el √≠ndice de su pel√≠cula favorita. Puede votar tantas veces como desee hasta que escriba 'salir'. Esta parte del c√≥digo demuestra c√≥mo recoger y procesar entradas del usuario en un bucle, actualizando el diccionario de votos seg√∫n las elecciones.

- **An√°lisis y Resultado Final:** Una vez que el usuario termina de votar, el programa utiliza reduce para determinar cu√°l pel√≠cula recibi√≥ m√°s votos, mostrando el resultado. Esto demuestra la aplicaci√≥n pr√°ctica de reduce para sintetizar una colecci√≥n de datos en un solo resultado.




```
"Titanic",
"El Padrino",
"Forrest Gump",
"La lista de Schindler",
"El Se√±or de los Anillos: El retorno del Rey"
```



````python
# Listado de pel√≠culas disponibles
peliculas =  [
    "Titanic",
    "El Padrino",
    "Forrest Gump",
    "La lista de Schindler",
    "El Se√±or de los Anillos: El retorno del Rey"
]
peliculas
````

````python
# --- 1. Definici√≥n de la Lista de Pel√≠culas y Contadores ---
PELICULAS = [
    "Titanic",
    "El Padrino",
    "Forrest Gump",
    "La lista de Schindler",
    "El Se√±or de los Anillos: El retorno del Rey"
]

# Inicializamos el contador de votos
votos = {pelicula: 0 for pelicula in PELICULAS}


# --- 2. Listado de Pel√≠culas (Usando enumerate) ---
def mostrar_peliculas():
    """Muestra la lista de pel√≠culas disponibles para la votaci√≥n."""
    print("-"*65)
    print("üé¨ ¬°Bienvenido a la Noche de Cine! Por favor, vota por tu pel√≠cula favorita de la siguiente lista.")
    print(("-"*65), "\n")
    for i, pelicula in enumerate(PELICULAS):
        print(f"[{i}] {pelicula}")
    print("\n", ("-"*65))


# --- 3. Registrar Votos en un Bucle ---
def registrar_votos(votos_actuales):
    """Permite al usuario votar en un bucle hasta que escribe 'salir'."""
    print("PARA VOTAR: introduce el n√∫mero (√≠ndice) de la pel√≠cula que se muestra en la lista anterior.")
    print("PARA TERMINAR LA VOTACI√ìN: escribe 'salir'.")

    while True:
        entrada = input("Tu voto (escribe √≠ndice o 'salir'): ").strip().lower()

        if entrada == 'salir':
            print("-"*65)
            print("üó≥Ô∏è Votaci√≥n finalizada.")
            print("-"*65)
            break

        try:
            indice_voto = int(entrada)

            if 0 <= indice_voto < len(PELICULAS):
                pelicula_votada = PELICULAS[indice_voto]
                votos_actuales[pelicula_votada] += 1
                print(f"‚úÖ Voto registrado para: {pelicula_votada}")
            else:
                print(f"‚ùå Error: El √≠ndice {indice_voto} no es v√°lido. Int√©ntalo de nuevo.")

        except ValueError:
            print("‚ùå Entrada inv√°lida. Por favor, introduce un n√∫mero de √≠ndice o 'salir'.")

    return votos_actuales


# --- 4. An√°lisis y Resultado Final ---
def analizar_votos(resultados_votos):
    """
    Analiza el diccionario de votos para encontrar la pel√≠cula con m√°s votos
    utilizando un bucle for.
    """
    if not any(resultados_votos.values()):
        return "\nNadie vot√≥. No hay pel√≠cula ganadora.\n"

    pelicula_ganadora = None
    max_votos = -1

    # Bucle para encontrar la pel√≠cula con el m√°ximo de votos
    for pelicula, votos in resultados_votos.items():
        if votos > max_votos:
            max_votos = votos
            pelicula_ganadora = pelicula

    # --- Manejo de Empates ---
    empates = [pelicula for pelicula, votos in resultados_votos.items() if votos == max_votos]

    if len(empates) > 1:
        nombres_empate = ", ".join(empates)
        return (f"\nüèÜ ¬°EMPATE! Las pel√≠culas m√°s populares son: {nombres_empate}\n "
                f"con {max_votos} votos cada una.")
    else:
        return (f"\nü•á ¬°LA GANADORA ES: {pelicula_ganadora}! Recibi√≥ un total de {max_votos} votos.\n")


# --- 5. Ejecuci√≥n Principal del Programa ---
if __name__ == "__main__":
    mostrar_peliculas()

    votos_finales = registrar_votos(votos)

    print("-"*65)
    print("--- RESULTADO FINAL DE LA VOTACI√ìN ---")
    print("-"*65)
    resultado = analizar_votos(votos_finales)
    print(resultado)

    print("-"*65)
    print("--- DETALLE DE LOS VOTOS ---")
    print(("-"*65), "\n")
    for pelicula, conteo in votos_finales.items():
        print(f"  {pelicula}: {conteo} votos")
````
