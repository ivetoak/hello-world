# Argumentos

```python
# ¿Cuando una función admite argumentos?
## parámetros = elementos dentro del parántesis de la función
## argumentos = valor que le damos a los elementos, se asignan cuando llamamos a la función

def suma_simple(num1, num2):
  return num1 + num2
```

```python
# Definimios variable + añadimos los argumentos a una función definida anteriormente tanto el numero de argumentos como la
# operación a realizar

resultado_suma = suma_simple(5, 5)
print(resultado_suma)
```

```python
# Los argumentos también pueden ser str, no solo int o float. Otra manera de concatenar caracteres.

resultado_suma = suma_simple("ho", "la")
print(resultado_suma)
```

```python
# Hay 2 tipos de argumentos especiales: *args & **kwargs
## *args recibe cualquier tipo de datos que yo le pase porque pongo el *. Este indica que espera varios valores/argumentos.

def funcion_ejemplo(*args):
  print("Argumentos posicionales (args):", args)

  # print("Argumentos clave-valor (kwargs):", kwargs)
```

```python
funcion_ejemplo(nombre='John', edad=25)
```

```python
# Hay 2 tipos de argumentos especiales: *args & **kwargs
## **kwargs = que *args pero las variables son clave=valor, como en un diccionario (k de 'key').

def funcion_ejemplo(**kwargs):
  # print("Argumentos posicionales (args):", args)

  print("Argumentos clave-valor (kwargs):", kwargs)
```

```python
funcion_ejemplo(nombre="John", edad=25)
```

```python
# Comparación de cuando ambas variables deben ser verdaderas.
## Si las 2 son verdaderas devuelve True, si no, False

a = True
b = True
resultado_and = a and b
resultado_and
```

```python
# Comparación de cuando + de 2 variables deben ser verdaderas. Si las 3 son verdaderas devuelve True, si no, False
## definido en línea 63 mayor_que en el que 5 es mayor a 3 y la función mayor_que devuleve previamente True

a = True
b = True
resultado_and = a and b and mayor_que
resultado_and
```

```python
## Cuando una o otra es verdadera

a = True
b = False
resultado_or = a or b
resultado_or
```

```python
## Cuando ninguna de las 2 es verdadera devuelve False

a = False
b = False
resultado_or = a or b
resultado_or
```

```python
# Cuando cualquiera de las variables mencionadas es verdadera.
## En este caso mayor_que es la verdadera y entonces devuelve True

a = True
b = False
resultado_or = a or b or mayor_que
resultado_or
```

*   de Pertenencia:

```python
# Cuando queremos ver si un valor 'X' pertenece a una lista predefinidia.
## En este caso 3 pertenece a la lista. Devuelve True.

lista = [1, 2, 3, 4, 5]
pertenece = 3 in lista
pertenece
```

```python
## En este caso 6 no pertenece a la lista. Devuelve False.

lista = [1, 2, 3, 4, 5]
pertenece = 6 in lista
pertenece
```

```python
# Cuando queremos ver si un valor 'X' NO pertenece a una lista predefinidia.
## En este caso 6 NO pertenece y devuelve True.

lista = [1, 2, 3, 4, 5]
no_pertenece = 6 not in lista
no_pertenece
```

```python
## En este caso 3 SÍ pertenece y devuelve False.

lista = [1, 2, 3, 4, 5]
no_pertenece = 3 not in lista
no_pertenece
```
